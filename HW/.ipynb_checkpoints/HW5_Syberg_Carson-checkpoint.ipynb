**Part D**: Suppose we wish to approach the question: is the flowering date becoming *earlier*?  One way to do this might be to generate confidence intervals comparing e.g. the last $n$ years of data to all the years that came before them and ask if they're coming from distributions with the same *means*.

Suppose we take the most recent 12 years of take, from 2010-2021.  What is the mean full-flowering date for those years?  Perform a hypothesis test at $\alpha=0.01$ testing whether or not the mean from this interval is the same as the mean for the prior all-observed years.

Write a sentence interpreting this result.

**Note:** there are a lot of missing observations, especially for centuries ago. Make sure you're properly not including these in any variance/confidence/etc. calculations.

```python
# need mean and variances of (first few hundred years) and (last 12 years)
earlyYears = df.loc[df['AD'] < 2010]
lateYears = df.loc[df['AD'] >= 2010]

# get first few hundred years mean and standard deviation
earlyMean = earlyYears['Full-flowering date (DOY)'].mean()
earlySize = earlyYears['Full-flowering date (DOY)'].count()
earlySTD = earlyYears['Full-flowering date (DOY)'].std()
earlyVar = earlySTD**2;

lateMean = lateYears['Full-flowering date (DOY)'].mean()
lateSize = lateYears['Full-flowering date (DOY)'].count()
lateSTD = lateYears['Full-flowering date (DOY)'].std()
lateVar = earlySTD**2;

alpha = 0.01
z_alpha = stats.norm.ppf(1-alpha/2)

tdof=(lateVar/lateSize + earlyVar/earlySize)**2/\
((lateVar/lateSize)**2/(lateSize-1)+(earlyVar/earlySize)**2/(earlySize-1))
t_alpha = stats.t.ppf(.995, df=tdof)
alpha = 0.01
meanDiff = lateMean - earlyMean;
pooledSE = np.sqrt(earlySTD**2/earlySize + lateSTD**2/lateSize);
t_stat = (meanDiff) / pooledSE;
print("present mean: ", lateMean, " compared to early mean: ", earlyMean);
print("CI: ", meanDiff - t_alpha*)
print("t_stat: ", t_stat, " compared to -t_alpha,v: ", -t_alpha)
```

$H_0$: $\mu = \mu_0$ in this case, $\mu$ is the true mean of the last 12 years full flowering date (DOY) and $\mu_0$ is the true mean of the last few hundred years before 2010.

$H_a$: $\mu < \mu_0$ meaning the true mean of the last 12 years (DOY) is smaller than the last few hundred years, or that the full flowering date has decreased.

$H_0$ can also be described as $\mu - \mu_0 = 0$

For this alternative hypothesis, to reject $H_0$ it was required that $t_{stat} < -t_{\alpha,v}$ and based off the calculations in the code, this seems to be the case. Therefore, we reject the null hypothesis and can assume that the full flowering date has indeed decreased in recent years.

- For *each* one of those 12-year intervals, generate the 99% confidence interval for the difference in the mean `FuFD`  over  those 12 years' to the mean from most recent 12 years from 2010-2021.

From that list of confidence intervals, print out a list of every starting year for an interval that contains zero. (e.g if the confidence interval for comparing 1618-1629 to 2010-2021 contains zero, then print `1618`.  Report:

```python
currMinYear = 1599
currMaxYear = 1612
CI_List = []
yearList = []
while(currMaxYear < 2010):
    currYearsA = df.loc[df['AD'] < currMaxYear]
    currYears = currYearsA.loc[df['AD'] > currMinYear]
    currMean = currYears['Full-flowering date (DOY)'].mean()
    currSize = currYears['Full-flowering date (DOY)'].count()
    currSTD = currYears['Full-flowering date (DOY)'].std()
    CI = [(lateMean - currMean) - t_alpha*(np.sqrt(currSTD**2/currSize + lateSTD**2/lateSize)), (lateMean - currMean) + t_alpha*(np.sqrt(currSTD**2/currSize + lateSTD**2/lateSize))]
    CI_List.append(CI)
    yearList.append(currMinYear + 1)
    currMinYear += 1
    currMaxYear += 1

count = 0
for i in CI_List:
    # determine if 0 in confidence interval
    if(0 > i[0] and 0 < i[1]):
        print(yearList[count])
    count += 1
```

### [15 points] Problem 3 - Nap time?


It is estimated that [about 75% of adults](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4997286/) in the United States drink coffee. Often, coffee is used to replace the need for sleep. It works alright, or so we think. Let's find out, in this exciting homework problem!

[One recent study](https://www.sciencedirect.com/science/article/pii/S1388245703002554) investigated the effects of drinking coffee, taking a nap, and having a ["coffee-nap"](https://lifehacker.com/naps-vs-coffee-which-is-better-when-youre-exhausted-1730643671) - the practice of drinking some coffee *and then* having a short nap. The study broke participants up into three groups of 15 participants each, where the groups would have a nap, or have a coffee, or have a coffee-nap, then perform a task where their reaction time was measured. In previous experiments the mean reaction time measurement was found to be normally distributed. The reaction time means (milliseconds, ms) and standard deviations for the three groups of participants are given in the table below.

$$
\begin{array}{c|c|c|c}
\textrm{Group} & \textrm{Sample Size} & \textrm{Mean} & \textrm{Standard Deviation} \\
\hline 
\textrm{Coffee+Nap} & 15 & 451.3 & 31.9 \\ 
\textrm{Coffee} & 15 & 494.2 & 39.6 \\ 
\textrm{Nap} & 15 & 492.8 & 45.2 \\ 
\end{array}
$$

**Part A**: Compute a 97.5% t-confidence interval for the mean reaction time measurement for participants in each of these three groups. (You should find three separate confidence intervals.) Report the results.

1. Can you make any conclusions regarding whether coffee, naps or both (coffee-naps) are better for faster reaction times?
2. Why did we use a t-distribution?

t-confidence interval is $[\bar{X} - t_{\alpha/2}\frac{s}{\sqrt{n}}, \bar{X} + t_{\alpha/2}\frac{s}{\sqrt{n}}]$, $t_{\alpha/2} = t_{0.0125} = 2.51$ calculated below.

Coffe+Nap CI: [ 430.6298196583792 ,  471.97018034162085 ]

Coffe CI: [ 468.54046578281555 ,  519.8595342171844 ]

Nap CI: [ 463.5118447824056 ,  522.0881552175944 ]

1. I cannot make any conclusions here because the confidence intervals overlap.
2. We used a t-distribution because of the small sample sizes in the data.

```python
# finding t_alpha/2
alpha = 0.025
t_alpha = stats.t.ppf(1-alpha/2, 14)
print("t_alpha/2: ", t_alpha)
print("Coffe+Nap CI: [", 451.3 - t_alpha*(31.9/np.sqrt(15)), ', ', 451.3 + t_alpha*(31.9/np.sqrt(15)), "]")
print("Coffe CI: [", 494.2 - t_alpha*(39.6/np.sqrt(15)), ', ', 494.2 + t_alpha*(39.6/np.sqrt(15)), "]")
print("Nap CI: [", 492.8 - t_alpha*(45.2/np.sqrt(15)), ', ', 492.8 + t_alpha*(45.2/np.sqrt(15)), "]")
```

**Part B**: Use an appropriate hypothesis test to determine if there sufficient evidence, at the $\alpha = 0.025$ significance level, to conclude that taking a nap promotes faster reaction time than drinking coffee.  Be sure to clearly explain the test that you're doing and state all hypotheses. Do all computations in Python, and report results.

```python
alpha = 0.025
t_alpha = stats.t.ppf(1-alpha/2, 14)
Xbar = 492.8 # sample mean time after nap
mu0 = 494.2 # mean time after coffee
s = 45.2 # sample standard deviation after nap
n = 15 # sample size

t = (Xbar - mu0) / (s/np.sqrt(n))

if(t > t_alpha):
    print("Reject H_0, Nap better than Coffee: ", t, ">", t_alpha)
else:
    print("Fail to reject H_0, Insufficient Evidence: t =", t, "!> t_alpha =", t_alpha)
```

Using test for population mean (small sample)\
$H_0:$ mean reaction time after nap = mean reaction time after coffee\
$H_a:$ mean reaction time after nap > mean reaction time after coffee

true mean reaction time after nap = $\mu$, true mean reaction time after coffee = $\mu_0$

Test statistic value: $t = \frac{\bar{X} - \mu_0}{s\div\sqrt{n}}$

In order to reject null hypothesis and find that nap does indeed promote faster reaction time than coffee, we need $\mu > \mu_0$ and $t > t_\alpha$

**Part D**: Generally speaking, we ust $\alpha$ to denote the "probability of error if the null hypothesis is true."  In parts B-C, we performed three hypothesis tests.  Suppost that each of these tests is **independent**, with underlying Probability of type I error of $\alpha=0.025$.  What's the probability that we committed *at least one* Type I error over those 3 tests?  Why might this be important?

If each of these tests is independent, and each had a probability of type I error of $\alpha = 0.025$, then the probability that we commited *at least one* Type I error over the 3 tests is equal to 1 minus the probability that we commited zero Type I errors over the 3 tests, or $1 - (1-\alpha)^3 = 1 - (1-0.025)^3 = 1 - 0.927 = 0.0731$ 

This might be important because it could tell us how likely it is that something observed in our data is actually true even though we rejected it, so it could tell us whether or not it's worth it to take another look at the data and do more testing to try and find out if one of our alternative hypotheses was viable.
